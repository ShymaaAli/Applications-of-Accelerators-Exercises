```python
import numpy as np
import matplotlib.pyplot as plt
#importing the data
datatable=np.genfromtxt('2016-07-11_ipm_data.txt', delimiter='e+0')
data=datatable[:,0]*10 
n=len(data)
# Maximum value calculation
maximum=0
for i in range(n):
    if data [i]>maximum:
        maximum=data [i]
        index=i
#plot the data
plt.plot(data)
plt.xlabel('position')
plt.ylabel('signal')
plt.title('ionisation beam')
#color the grid in black
plt.grid(color='k') 
plt.plot(index, maximum, marker='o', mfc='r') #add red dot at maximum
#Save image as png
plt.savefig('shymaa_Ex3_b.png') 
```
